**Import the relevant python libraries for the analysis**

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from google.colab import files
import seaborn as sns
uploaded = files.upload()
```

```python
data = pd.read_csv('Contact Information (Responses) - Form Responses 1.csv')
```

```python
data.drop(["Timestamp","Email Address"],axis = 1)
```

```python
# @title How often do you exercise in a week? vs Do you feel that regular exercise impacts your academic performance? How much on scale from 1 to 5 ? If five is the best and 0 is doesn't affect me at all?

data.plot(kind='scatter', x='How often do you exercise in a week?', y="Do you feel that regular exercise impacts your academic performance? How much on scale from 1 to 5 ? If five is the best and 0 is doesn't affect me at all?", s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#@title PEARSON'S R
data['How often do you exercise in a week?'].corr(data["Do you feel that regular exercise impacts your academic performance? How much on scale from 1 to 5 ? If five is the best and 0 is doesn't affect me at all?"])
```

```python
#@title How often do you exercise in a week?(mean , median and mode)
dat = data['How often do you exercise in a week?']

mean_value = dat.mean()
median_value = dat.median()
mode_value = dat.mode()[0]

plt.hist(dat, bins=30, color=sns.palettes.mpl_palette('Dark2')[0])

plt.axvline(x=mean_value, color='r', linestyle='dashed', linewidth=4)
plt.axvline(x=median_value, color='black', linestyle='dashed', linewidth=4)
plt.axvline(x=mode_value, color='b', linestyle='dashed', linewidth=4)

plt.legend(['Mean', 'Median', 'Mode'])

plt.gca().spines[['top', 'right']].set_visible(True)

plt.title('Histogram of How often do you exercise in a week?')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
#@title Box Plot of How often do you exercise in a week?
column = 'How often do you exercise in a week?'


Q1 = data["How often do you exercise in a week?"].quantile(0.25)
Q3 = data["How often do you exercise in a week?"].quantile(0.75)
median = data["How often do you exercise in a week?"].median()

plt.boxplot(data["How often do you exercise in a week?"])
plt.title('Box Plot of ' + "How often do you exercise in a week?")

plt.axhline(Q1, color='r', linestyle='dashed', linewidth=2)
plt.axhline(Q3, color='b', linestyle='dashed', linewidth=2)
plt.axhline(median, color='g', linestyle='dashed', linewidth=2)

plt.show()
```
